# 7. Metrics

```python
from pathlib import Path
import shutil
import re

noisy_root = Path("/kaggle/working/Evaluation/test/noisy")
clean_root = Path("/kaggle/working/Evaluation/test/clean")
clean_aligned_root = Path("/kaggle/working/Evaluation/test/clean_aligned")

if clean_aligned_root.exists():
    shutil.rmtree(clean_aligned_root)
clean_aligned_root.mkdir(parents=True, exist_ok=True)

exts = {".wav", ".flac"}

# indeks clean po: (speaker, utt_id), np. (p104, 00002)
clean_idx = {}
for cp in clean_root.rglob("*"):
    if cp.is_file() and cp.suffix.lower() in exts:
        rel = cp.relative_to(clean_root)
        speaker = rel.parts[0] if len(rel.parts) > 1 else ""
        utt_id = cp.stem  # np. 00002
        clean_idx[(speaker, utt_id)] = cp

mapped = 0
missing = []

for np in noisy_root.rglob("*"):
    if not (np.is_file() and np.suffix.lower() in exts):
        continue

    rel_noisy = np.relative_to(noisy_root)
    speaker = rel_noisy.parts[0] if len(rel_noisy.parts) > 1 else ""
    noisy_stem = np.stem  # np. 00002_-1.7dB

    # wyciągnij ID utterance z noisy: "00002" z "00002_-1.7dB"
    utt_id = re.split(r"[_-]", noisy_stem, maxsplit=1)[0]

    cp = clean_idx.get((speaker, utt_id))
    if cp is None:
        missing.append(str(rel_noisy))
        continue

    out_path = clean_aligned_root / rel_noisy
    out_path.parent.mkdir(parents=True, exist_ok=True)
    shutil.copy2(cp, out_path)
    mapped += 1

print("Mapped clean files:", mapped)
print("Missing:", len(missing))
if missing[:10]:
    print("Przykładowe brakujące:")
    for x in missing[:10]:
        print("-", x)
```
